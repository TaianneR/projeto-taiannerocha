```python



df_mundial = pd.read_csv('mundial.csv')


print(df_mundial.head())


print(df_mundial.describe())
```

```python
correlacao_operacional_pesquisa = df_mundial['Operating Environment'].corr(df_mundial['Research'])
correlacao_operacional_desenvolvimento = df_mundial['Operating Environment'].corr(df_mundial['Development'])

print(f'Correlação Ambiente Operacional e Pesquisa em IA: {correlacao_operacional_pesquisa}')
print(f'Correlação Ambiente Operacional e Desenvolvimento em IA: {correlacao_operacional_desenvolvimento}')
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt


df_renda = df_mundial[['Country', 'Income group', 'Total score']]

plt.figure(figsize=(12, 8))
sns.barplot(x='Income group', y='Total score', data=df_renda.sort_values(by='Total score', ascending=False))
plt.title('Escore Total de IA por Grupo de Renda')
plt.xlabel('Grupo de Renda')
plt.ylabel('Escore Total de IA')
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt



# Agrupe os dados por país e calcule a média da pontuação de talento
talent_por_pais = df_mundial.groupby('Country')['Talent'].mean().sort_values(ascending=False)


top5_paises_talento = talent_por_pais.head(5)

plt.figure(figsize=(10, 6))
sns.barplot(x=top5_paises_talento.index, y=top5_paises_talento.values, palette='viridis')
plt.title('Top 5 Países com Maior Pontuação de Talent no Índice Global de IA')
plt.xlabel('País')
plt.ylabel('Média de Pontuação de Talent')
plt.show()
```
